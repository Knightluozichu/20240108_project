- 微分学
    - 什么是微分学？
        - 它主要研究的是函数在某一点的局部性质，尤其是函数的瞬时变化率，即导数。
        - 微分学的核心概念是导数，它描述了一个量相对于另一个量的变化速度。
        - 微分学通过引入导数的概念，为研究函数的切线斜率、曲线的凹凸性、函数的极值等问题提供了数学工具。
    - 它解决了什么问题？
        - 变化率问题
        - 最优化问题
        - 物理定律的精确描述
        - 模型建立和分析
        - 误差分析
    - 导数
        - 导数描述的是函数在某一点处的瞬时变化率，它是通过极限的概念来精确定义的。通过计算导数，我们可以了解函数在特定点的局部行为，比如速度、加速度、成本变化等。
        - f'(a) = lim(h->0) [f(a+h) - f(a)] / h
    - 导函数 导函数是原函数在某一点处导数的集合
        - 常数法则：如果f(x)是一个常数，那么它的导数是0。
        - 幂法则：如果f(x) = x^n（n是一个实数），那么f'(x) = n * x^(n-1)。
        - 乘积法则：如果f(x)是两个函数的乘积，比如f(x) = g(x) * h(x)，那么f'(x) = g'(x) * h(x) + g(x) * h'(x)。
        - 商法则：如果f(x)是两个函数的商，比如f(x) = g(x) / h(x)，那么f'(x) = (h(x) * g'(x) - g(x) * h'(x)) / h(x)^2，前提是分母h(x)不等于0。
        - 链式法则：如果f(x)是另一个函数的复合，比如f(x) = g(h(x))，那么f'(x) = g'(h(x)) * h'(x)。

### 微积分概述

#### 1. 导数
- **导数的定义**: 描述函数在某一点处的瞬时变化率。
- **导函数**: 一个函数的导数构成的新的函数。
- **导函数的计算**: 通过一定的数学规则和方法来求解原函数的导函数。
- **复合函数的求导法则**: 用于求解由多个函数复合而成的函数的导数。
- **费马定理**: 确定函数极值点的一个重要定理，指出极值点处的导数为零。

#### 2. 偏导数
- **多元函数的偏导数**: 对于具有多个自变量的函数，偏导数描述了函数沿某一特定变量的变化率，而其他变量保持不变。
  - **以二元函数为例**: 对于函数z=f(x,y)，可以分别对x和y求偏导数。
  - **对x的偏导数**: 表示函数对x的变化敏感度。
  - **对y的偏导数**: 表示函数对y的变化敏感度。

#### 3. 偏导函数
- **偏导数的意义**: 偏导数在多元函数分析中具有重要的作用，特别是在优化问题、物理模型和机器学习算法中。

#### 4. 泰勒展式
- **泰勒展式的定义**: 利用函数在某一点的各阶导数值，构建一个多项式函数来近似原函数。
- **泰勒展式的用途**: 用于近似计算函数在某一点的值，特别是当直接计算复杂或不可能时。
- **多项式逼近**: 通过泰勒展式，可以用一个多项式函数去逼近一个给定的函数，从而在某个邻域内得到较为精确的结果。
